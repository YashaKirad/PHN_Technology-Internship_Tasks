```python
from sklearn import datasets
from sklearn import metrics
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

     
```

```python
url = "https://raw.github.com/mattdelhey/kaggle-titanic/master/Data/train.csv"
df = pd.read_csv(url)
df
```

```python
# Find the names of passengers younger than 35 years
young_passengers = df[df['age'] < 35]['name']
print("the names of passengers younger than 35 years.")
print(young_passengers)

# Print the rows from index 10 to 25 and columns 3 to 5
print("rows from index 10 to 25 and columns 3 to 5")
subset = df.iloc[10:26, 3:6]
print(subset)

# Find the statistics aggregate of Age & Fare using the DataFrame.agg() method
statistics = df[['age', 'fare']].agg(['min', 'max', 'mean', 'median', 'std'])
print("the statistics aggregate of Age & Fare using the DataFrame.agg() method")
print(statistics)

# Find the mean ticket fare price for each sex and cabin class combination
fare_mean = df.groupby(['sex', 'pclass'])['fare'].mean()
print("the mean ticket fare price for each of the sex and cabin class combinations")
print(fare_mean)
```
