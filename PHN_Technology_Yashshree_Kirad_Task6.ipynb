```python
import numpy as np

def generate_linear_equations_dataset(num_equations, min_x, max_x, min_slope, max_slope, min_intercept, max_intercept):
    dataset = []
    
    for _ in range(num_equations):
        # Randomly generate slope, intercept, and input variables
        slope = np.random.uniform(min_slope, max_slope)
        intercept = np.random.uniform(min_intercept, max_intercept)
        x1 = np.random.uniform(min_x, max_x)
        x2 = np.random.uniform(min_x, max_x)
        
        # Calculate the corresponding output value
        y = slope * x1 + intercept * x2
        
        # Append the equation to the dataset
        equation = {'x1': x1, 'x2': x2, 'y': y}
        dataset.append(equation)
    
    return dataset

# Example usage
dataset = generate_linear_equations_dataset(num_equations=100, min_x=0, max_x=10, min_slope=-5, max_slope=5, min_intercept=-10, max_intercept=10)

# Print the generated dataset in CSV-like format
print("{:<8} {:<8} {:<8}".format("x1", "x2", "y"))
print("--------------------")
for equation in dataset:
    print("{:<8.2f} {:<8.2f} {:<8.2f}".format(equation['x1'], equation['x2'], equation['y']))
```

```python
X = np.array([[equation['x1'], equation['x2']] for equation in dataset])
y = np.array([equation['y'] for equation in dataset])

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Perform data normalization on input variables and output values
X_normalized = scaler.fit_transform(X)
y_normalized = scaler.fit_transform(y.reshape(-1, 1))

# Print the normalized dataset with proper headings and spacing
print("{:<15} {:<15} {:<15}".format("x1_normalized", "x2_normalized", "y_normalized"))
print("--------------------------------------------")
for i in range(len(dataset)):
    equation = dataset[i]
    x1_normalized = X_normalized[i, 0]
    x2_normalized = X_normalized[i, 1]
    y_normalized_value = y_normalized[i, 0]
    print("{:<15.4f} {:<15.4f} {:<15.4f}".format(x1_normalized, x2_normalized, y_normalized_value))
```

```python
# Extract input variables and output values from the dataset
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

X = np.array([[equation['x1'], equation['x2'], equation['x1'] * equation['x2']] for equation in dataset])
y = np.array([equation['y'] for equation in dataset])

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Perform data normalization on input variables and output values
X_normalized = scaler.fit_transform(X)
y_normalized = scaler.fit_transform(y.reshape(-1, 1))

# Split the dataset into training and testing sets (you may adjust the split ratio as per your requirements)
split_ratio = 0.8
split_index = int(len(dataset) * split_ratio)
X_train = X_normalized[:split_index]
y_train = y_normalized[:split_index]
X_test = X_normalized[split_index:]
y_test = y_normalized[split_index:]

# Train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)

# Inverse transform the normalized predictions and actual values
y_pred_inverse = scaler.inverse_transform(y_pred.reshape(-1, 1))
y_test_inverse = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calculate evaluation metrics
mae = mean_absolute_error(y_test_inverse, y_pred_inverse)
mse = mean_squared_error(y_test_inverse, y_pred_inverse)
r2 = r2_score(y_test_inverse, y_pred_inverse)

# Print the evaluation metrics
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("R-squared Score (R2):", r2)
```
