```python
import openpyxl

# Create a new Excel workbook
workbook = openpyxl.Workbook()
sheet = workbook.active

# Add the header row
header = ['Fruit', 'Quantity']
sheet.append(header)

# Add the data rows
data = [
    ['Apple', 10],
    ['Banana', 5],
    ['Apple', 15],
    ['Orange', 20],
    ['Banana', 10],
    ['Grapes', 8],
    ['Mango', 12],
    ['Kiwi', 6],
    ['Pineapple', 9]
]

for row in data:
    sheet.append(row)

# Save the Excel spreadsheet
filename = 'fruit_data.xlsx'
workbook.save(filename)

print("Excel spreadsheet created successfully.")
```

```python
import openpyxl

# Read the data from the original Excel spreadsheet
original_workbook = openpyxl.load_workbook('fruit_data.xlsx')
original_sheet = original_workbook.active

# Display the original Excel file
print("Original Excel file:")
for row in original_sheet.iter_rows(values_only=True):
    print(row)

# Create a dictionary to store the total quantity of each fruit
fruit_totals = {}

# Iterate through the rows in the original spreadsheet (excluding the header row)
for row in original_sheet.iter_rows(min_row=2, values_only=True):
    fruit = row[0]
    quantity = row[1]

    # Update the total quantity for the fruit in the dictionary
    if fruit in fruit_totals:
        fruit_totals[fruit] += quantity
    else:
        fruit_totals[fruit] = quantity

# Create a new Excel workbook and sheet for the results
result_workbook = openpyxl.Workbook()
result_sheet = result_workbook.active

# Write the results to the new spreadsheet
result_sheet.cell(row=1, column=1, value="Fruit")
result_sheet.cell(row=1, column=2, value="Total Quantity")

# Iterate through the fruit_totals dictionary and write the fruit name and total quantity to the new spreadsheet
row_num = 2
for fruit, total_quantity in fruit_totals.items():
    result_sheet.cell(row=row_num, column=1, value=fruit)
    result_sheet.cell(row=row_num, column=2, value=total_quantity)
    row_num += 1

# Save the new spreadsheet
result_file = "fruit_totals.xlsx"
result_workbook.save(result_file)

# Display the updated Excel file
print("\nUpdated Excel file:")
for row in result_sheet.iter_rows(values_only=True):
    print(row)

print("\nResults saved successfully to", result_file)
```
